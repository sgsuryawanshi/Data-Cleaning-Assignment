```python
def print_anagrams(anagrams):
    """Uses a generator to call and print 5 items from mydict"""
    fp = (fp for fp in anagrams)

    print ("Sample from anagram dict:")
    for i in range(1, 6):
        # call once, print twice
        fp_next = fp.next()
        print ("%s) %s:" % (i, fp_next), anagrams[fp_next])

        print ("\n")
```
